```python
import pandas as pd
import random

# 더미 데이터 생성 (50개)
dummy_data = {
    '속도': [random.uniform(3.0, 4.5) for _ in range(50)],
    '몸무게': [random.uniform(65, 90) for _ in range(50)],
    '키': [random.uniform(160, 190) for _ in range(50)],
    '민첩성': [random.randint(5, 10) for _ in range(50)],
    '밸런스': [random.randint(4, 9) for _ in range(50)],
    '골 수': [random.randint(0, 10) for _ in range(50)]
}

# 데이터프레임 생성
df_dummy = pd.DataFrame(dummy_data)

# 데이터 CSV 파일로 저장
df_dummy.to_csv('player_data.csv', index=False)
print("더미 데이터가 성공적으로 저장되었습니다.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Colab 한글 폰트 설정
plt.rc('font', family='NanumGothic')

# 저장된 CSV 파일 불러오기
df = pd.read_csv('player_data.csv')

# 데이터 분석
correlation_matrix = df.corr()

# 시각화
plt.figure(figsize=(10, 8))
# Heatmap 그리기
sns.heatmap(
    correlation_matrix,    # 상관 행렬 데이터
    annot=True,             # 각 셀에 숫자 표시
    cmap='coolwarm',        # 색상 맵
    linewidths=.5           # 셀 간 경계선 굵기
)
# 그래프 제목 설정
plt.title('능력치 및 골 수 간의 상관 관계')

# 그래프 표시
plt.show()
```
